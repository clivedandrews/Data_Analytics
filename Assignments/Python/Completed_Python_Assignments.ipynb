```python
# MODULE: 1) Write a Python program to calculate number of days between two dates. Hint: use Datetime package/module.
# Sample dates : (2014, 7, 2), (2014, 7, 11)
# Expected output : 9 days

def dateDiff(date1, date2):
    import datetime
    dt1 = datetime.date(*date1)
    dt2 = datetime.date(*date2)
    diffdays = (dt2 - dt1).days
    print(f"The difference between the two dates is {diffdays} days")
    

dateDiff((2014, 7, 2),(2014, 7, 11))
dateDiff((2023, 4, 2),(2023, 4, 29))
```

```python
# FUNCTIONS: 7) Write a Python program to get the Fibonacci series between 0 to 50.
   

def Fib_0_to_50():
    Fib_str = ""
    element = 0
    prev_element = 1
    while element <= 50:
        Fib_str = Fib_str + " " + str(element)
        tmp_val = element
        element = element + prev_element
        prev_element = tmp_val
    return(Fib_str)
    
print("The Fibonacci series between 0 and 50 is as follows -", Fib_0_to_50())
            
        
    
    
```

```python
# FUNCTIONS: 9) Write a Python function that takes a list and returns a new list with unique elements of the first list.
# Sample List : [1,2,3,3,3,3,4,5]
# Unique List : [1, 2, 3, 4, 5]

def uniqueList(list):
    unique_list = []
    for element in list:
        if (element not in unique_list):
            unique_list.append(element)
    return(unique_list)

listVal1 = [1,2,3,3,3,3,4,5]
print("The Unique List is as follows -", uniqueList(listVal1))

listVal2 = [5,5,6,8,2,5,8,3,3,9,0,2,6,1,2,1,5,7,3,3,4,9,2,2,4,5,3,0,1,2,3,3,3,3,4,5]
print("The Unique List is as follows -", uniqueList(listVal2))

listVal3 = [9,52,3,30,1,2,73,3,73,3,1,84,3,3,33,33,5,36,8,2,52,84,52]
print("The Unique List is as follows -", uniqueList(listVal3))
```

```python
# STRINGS: 1) Write a Python program to concatenate all elements in a list into a string and return it.

# Note: We are concatinating using single space i.e. " "
def listToString(myList):
    MyString = ""
    for ListElement in myList:
        MyString = MyString + " " + ListElement
    print(f"The string that is created using the list elements is - {MyString}")

    
myStringList1 = ["WayMaker", "Miracle Worker", "Promise Keeper", "Light in the Darkness", "That", "is", "who", "you", "are"]
listToString(myStringList1)


myListVals = ["1", "2.5", "i", "am"] # All elements need to be a string in order to be concatenated
listToString(myListVals)

```

```python
# Logical Operators
#-------------------

#----------------------------------------------------------------------

# The Logical Operator 'and'
# The 'and' operator returns True if both statements are true

print("\nTHE LOGICAL OPERATOR 'and'\n------------------------------\n")

Logical_Val = 7 # Initializing the variable 'Logical_Val'

# Condition
Logical_Val > 4 and Logical_Val < 15

# Here both conditions match as 7 is greater than 4 and also lesser than 15 so the result is 'True'
print(f"With Logical_Val as {Logical_Val}, the result for 'Logical_Val < 5 and Logical_Val < 1' is {Logical_Val > 4 and Logical_Val < 15}")

Logical_Val = 3 # Changing the value of the variable 'Logical_Val'

# Here the condition 3 is greater than 4 does not match although 3 is lesser than 15 and so the result is 'False'. 
# If one condition is True and the other is False then the result is False
print(f"With Logical_Val as {Logical_Val}, the result for 'Logical_Val < 5 and Logical_Val < 1' is {Logical_Val > 4 and Logical_Val < 15}")

#----------------------------------------------------------------------

# The Logical Operator 'or'
# The 'or' operator returns True if either one of the statements is true
# In other words, the 'or' operator returns false if both statements are false

print("\n\nTHE LOGICAL OPERATOR 'or'\n------------------------------\n")


# Condition
Logical_Val > 4 or Logical_Val < 15

Logical_Val = 7 # Initializing the variable 'Logical_Val'

# Here both conditions match as 7 is greater than 4 and also lesser than 15 so the result is 'True'
print(f"With Logical_Val as {Logical_Val}, the result for 'Logical_Val < 5 or Logical_Val < 1' is {Logical_Val > 4 or Logical_Val < 15}")

Logical_Val = 3 # Changing the value of the variable 'Logical_Val'

# Here the condition 3 is greater than 4 does not match although 3 is lesser than 15 and 
# since one of them is True, the outcome is True. 
print(f"With Logical_Val as {Logical_Val}, the result for 'Logical_Val < 5 or Logical_Val < 1' is {Logical_Val > 4 or Logical_Val < 15}")



# Condition
Logical_Val > 5 or Logical_Val > 10

# Updating the value of the variable 'Logical_Val'
Logical_Val = 2

# Here both conditions do not match i.e., 2 is not greater than 5 nor 10 so the outcome is False
print(f"With Logical_Val as {Logical_Val}, the result for 'Logical_Val > 5 or Logical_Val > 10' is {Logical_Val > 5 or Logical_Val > 10}")


#----------------------------------------------------------------------
# The Logical Operator 'not'
# The 'not' operator returns True if the condition returns False and vice-versa
#----------------------------------------------------------------------


print("\n\nTHE LOGICAL OPERATOR 'not'\n------------------------------\n")

# Condition
Logical_Val > 4 and Logical_Val < 15

# Initializing the variable 'Logical_Val'
Logical_Val = 2

# Here outcome of the condition 'Logical_Val > 4 and Logical_Val < 15' is False 
# but the result of the 'not' for this condition will be True
print(f"With Logical_Val as {Logical_Val}, the result for 'not(Logical_Val > 4 and Logical_Val < 15)' is {not(Logical_Val > 4 and Logical_Val < 15)}")

# Updating the variable 'Logical_Val'
Logical_Val = 7

# Here outcome of the condition 'Logical_Val > 4 and Logical_Val < 15' is True 
# but the result of the 'not' for this condition will be False
print(f"With Logical_Val as {Logical_Val}, the result for 'not(Logical_Val > 4 and Logical_Val < 15)' is {not(Logical_Val > 4 and Logical_Val < 15)}")

print("\n------------------------------\n")
```

```python
# Python Identity Operators
#--------------------------

var1 = ["apple", "banana"]
var2 = ["apple", "banana"]

var3 = var1

# Condition
var1 is var3

# Since var3 points to var1 and the outcome of the condition 'var1 is var3' or vice-versa i.e.
# 'var1 is var3' is True
print(f"The outcome of the condition 'var1 is var3' is {var1 is var3}")
# The vice-versa will also hold good
print(f"The outcome of the condition 'var3 is var1' is {var3 is var1}")

# Condition
var1 is var2

# Since var1 and var2 point to different locations although their value is the same, 
# the outcome of the condition 'var1 is var2' is False and so is vice-versa
print(f"The outcome of the condition 'var1 is var2' is {var1 is var2}")
print(f"The outcome of the condition 'var2 is var1' is {var2 is var1}")

#----------------------------------------------------------------------
```

```python
# Python Membership Operators
#----------------------------

# The Membership Operator 'in'
# If a particular element is within a list of elements then it will return true or else false
fruits = ["apple", "banana"]

print(f'The outcome of the condition \'"banana" in fruits\' is {"banana" in fruits}')
print(f'The outcome of the condition \'"pinapple" in fruits\' is {"pinapple" in fruits}')

# The Membership Operator 'not in'
# If a particular element is not within a list of elements then it will return true or else false
fruits = ["apple", "banana"]

print(f'The outcome of the condition \'"pinapple" not in fruits\' is {"pinapple" not in fruits}')
print(f'The outcome of the condition \'"apple" not in fruits\' is {"apple" not in fruits}')
```

```python
# Python Bitwise Operators
#----------------------------

#-----------------------------------------------------------------------------------------------------------

# The Bitwise Operator AND sybolized as '&'
# This Bitwise Operator AND sybolized as '&' sets each bit to 1 if both corresponding bits in the two numbers are 1

print("\n\nTHE BITWISE OPERATOR AND SYMBOLIZED AS '&'\n---------------------------------------------")

Var1 = 13 # Binary value is 0000000000001101
Var2 = 9  # Binary value is 0000000000001001

# The outcome of the operation 'Var1 & Var2' is 0000000000001001 i.e., 9
print(f"The outcome of the condition '{Var1} & {Var2}' is {Var1 & Var2}")

#-----------------------------------------------------------------------------------------------------------

# The Bitwise Operator OR sybolized as '|'
# This Bitwise Operator OR sybolized as '|' sets each bit to 1 if one of the two corresponding bits in the two numbers is 1

print("\n\nTHE BITWISE OPERATOR OR SYMBOLIZED AS '|'\n---------------------------------------------")

Var1 = 13 # Binary value is 0000000000001101
Var2 = 9  # Binary value is 0000000000001001

# The outcome of the operation 'Var1 | Var2' is 0000000000001101 i.e., 13
print(f"The outcome of the condition '{Var1} | {Var2}' is {Var1 | Var2}")

#-----------------------------------------------------------------------------------------------------------

print("\n\nTHE BITWISE OPERATOR XOR SYMBOLIZED AS '^'\n---------------------------------------------")

# The Bitwise Operator XOR sybolized as  '^'
# This Bitwise Operator XOR sybolized as '^' sets each bit to 1 if only one of the two corresponding bits in the two numbers is 1

Var1 = 13 # Binary value is 0000000000001101
Var2 = 9  # Binary value is 0000000000001001

# The outcome of the operation 'Var1 ^ Var2' is 0000000000000100 i.e., 4
print(f"The outcome of the condition '{Var1} ^ {Var2}' is {Var1 ^ Var2}")

#-----------------------------------------------------------------------------------------------------------

print("\n\nTHE BITWISE OPERATOR NOT SYMBOLIZED AS '~'\n---------------------------------------------")

# The Bitwise Operator NOT sybolized as  '~'
# This Bitwise Operator NOT sybolized as '~' inverts each bit (0 becomes 1 and 1 becomes 0)

Var = 13 # Binary value is 0000000000001101

# The outcome of the operation '~13' is 1111111111110010 i.e., -14
print(f"The outcome of the condition '~{Var}' is {~Var}")

#-----------------------------------------------------------------------------------------------------------

print("\n\nTHE BITWISE OPERATOR 'ZERO FILL LEFT SHIFT' SYMBOLIZED AS '<<'\n---------------------------------------------")

# The Bitwise Operator 'ZERO FILL LEFT SHIFT' sybolized as '<<'
# This Bitwise Operator 'ZERO FILL LEFT SHIFT' sybolized as '<<' 
# shifts bits to left by pushing zeros in from the right and lets the leftmost bits fall off

Var = 13 # Binary value is 0000000000001101

# The outcome of the operation '13 << 2' is 0000000000110100 i.e., -52
print(f"The outcome of the condition 'Var << 2' is {Var << 2}")

#-----------------------------------------------------------------------------------------------------------

print("\n\nTHE BITWISE OPERATOR 'SIGNED RIGHT SHIFT' SYMBOLIZED AS '>>'\n---------------------------------------------")

# The Bitwise Operator 'SIGNED RIGHT SHIFT' sybolized as '>>'
# This Bitwise Operator 'SIGNED RIGHT SHIFT' sybolized as '>>' 
# shifts bits to right by pushing zeros in from the left and lets the rightmost bits fall off

Var = 13 # Binary value is 0000000000001101

# The outcome of the operation '13 >> 2' is 0000000000000011 i.e., 3
print(f"The outcome of the condition 'Var >> 2' is {Var >> 2}")

#-----------------------------------------------------------------------------------------------------------
```

```python
# BASIC PROGRAMS: 6) Create a 3x3 matrix with values ranging from 2 to 10 using numpy

import numpy as np
import random

myList=[]
n = 9 # Since we need to create a 3x3 matrix we need 9 elements
for i in range(n):
    myList.append(random.randint(2,10))

# Creating the array using NumPy
myArray = np.array(myList)

# In order to change the shape of the 1 dimensional array to a 3x3 matrix, we use the reshape function in NumPy
myMatrix = myArray.reshape(3,3)
print(myMatrix)
```
